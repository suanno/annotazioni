```python
import numpy as np
import matplotlib.pyplot as plt
f, ax = plt.subplots()
N = 2.24*1000
#xlattice = (np.arange(0,N)-(N-1)/2)*a
xlattice = np.arange(0,N)
```

```python
with open('before.txt', 'r') as file:
    data = file.read().splitlines()
N = len(data)
psi = np.zeros(N, dtype=complex)
psi2 = np.zeros(N, dtype=float)
for i in range(0,N):
    z = data[i].split(" +j")
    I = float(z[0])
    Q = float(z[1])
    #data[i] = data[i] + "j"
    #print(data[i]);
    psi[i] = I + 1j*Q
    psi2[i] = I**2+Q**2
ax.plot(xlattice, psi2)
```
