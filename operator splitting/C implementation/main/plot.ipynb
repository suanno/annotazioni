```python
import numpy as np
import matplotlib.pyplot as plt
f, ax = plt.subplots()
```

```python
with open('before.txt', 'r') as file:
    before = file.read().splitlines()
with open('then.txt', 'r') as file:
    then = file.read().splitlines()
N = len(before)
psi2_before = np.zeros(N, dtype=float)
psi2_then = np.zeros(N, dtype=float)
for i in range(0,N):
    z = before[i].split(" +j")
    I = float(z[0])
    Q = float(z[1])
    psi2_before[i] = I**2+Q**2
    z = before[i].split(" +j")
    I = float(z[0])
    Q = float(z[1])
    psi2_then[i] = I**2+Q**2
ax.plot(psi2_before)
ax.plot(psi2_then)
```
