```python
# Generate random N dimensional complex array
# and write it into a .txt

import numpy as np
N = int(2048);
real = np.random.rand(N)
imag = np.random.rand(N)

# Write into txt
with open("input.txt", "w") as txt_file:
    txt_file.write(str(N)+"\n")
    for i in range(0,N):
        txt_file.write(str(round(real[i],5))+" "+str(round(imag[i],5))+"\n") # works with any number of elements in a line
```

```python
# DFT
def DFT(N, vec):
    vec_k = np.zeros(N, dtype=complex)
    for k in range(0,N):
        for n in range(0,N):
            vec_k[k] = vec_k[k] + vec[n]*np.exp(-1j*2*np.pi*k*n/N)
    return vec_k
```
