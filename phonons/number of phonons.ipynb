We calculate the (statistical average) number of phonons in the Debye approximation; at low and high temperatures.


At low temperatures the Debye approximation is a good approximation, while it is not at hight temperatures.

I wish to calculate one day the phonons number at hight temperatures using the real (sampled) density of phonons states and not the Debye approximation's one.

# Introducing Debye Model

### Finite number of oscillation modes (3N)
Normal modes of oscillations of a Bravais lattice (without basis) are
$$u_{r,\mathbf{k}}(\mathbf{x}) = \epsilon_r e^{i(\mathbf{k}\cdot\mathbf{x}-\omega_r(\mathbf{k})t)}$$
Where $\epsilon_1, \epsilon_2, \epsilon_3$ are 3 orthogonal versor that descbribe the polarization of the oscillation. 

Imposing _PBC_ boundary conditions $u_{r,\mathbf{k}}(\mathbf{x} + L) = u_{r,\mathbf{k}}(\mathbf{x})$, where $L = Na$ is the lenght of the crystal and $N, a$ are the number of atoms from side to side and the lattice spacing.
This requires that $\mathbf{k}$ can assume only the values $\mathbf{k} = \frac{2\pi}{L}\mathbf{n}$ with $\mathbf{n}\in\mathcal{Z^3}$.

Then we note that 
$$u_{r,(\mathbf{k} + \frac{2\pi}{L}N)}(\mathbf{x}) = \epsilon_r e^{i(\mathbf{k}\cdot\mathbf{x} + \frac{2\pi}{L}N\cdot\mathbf{x} -\omega_r(\mathbf{k})t)}$$
but $\frac{2\pi}{L}N = \frac{2\pi}{Na}N = \frac{2\pi}{a}$ and $\mathbf{x}$ is discretized as it can only assume values corresponding to lattice sites, so $\mathbf{x} = ma$ with $m\in\mathcal{Z}$.

As a consequence, $\frac{2\pi}{L}N$ is always a multiple of $2\pi$ and, as it shows up in the phase of the complex exponential, we conclude that 
$$u_{r,(\mathbf{k} + \frac{2\pi}{L}N)} = u_{r,\mathbf{k}}$$
meaning that they are _the same oscillation mode_.

So $\mathbf{k} = \frac{2\pi}{L}\mathbf{n}$ with $\mathbf{n}\in\mathcal{Z^3}$ **and** $n_x, n_y, n_z < N$. And remember that $r\in\{1,2,3\}$.

This means that the (different) normal modes of oscillations in a Bravais lattice (without basis) are **3N**.

### The Debye approximation
The number of oscillation modes is $3N$, in particular there are $N$ modes for each polarization $r = 1,2,3$ and the corresponding wave vectors $k$ are $N$ equispaced discrete points in the _fisrt Brouillen zone_ $k_i = \frac{2\pi}{L}n_i$ with $i = x,y,z$ and $n_i = 0, 1, ..., N-1$.

The _phonon dispersion relations_ $\omega_r(\mathbf{k})$ are three functions ($r = 1, 2, 3$) that depend on the crystal shape and composition.
For normal modes (that we will call _phonon states_ later) with momentum $k\simeq 0$, all the three dispersion relations are linear
$$\omega_r(\mathbf{k}) = c_rk + o(k^2)\quad \text{if}\quad k\rightarrow 0$$
where $k = |\mathbf{k}|$ and the three values $c_r$ are numbers (in principle different).

The **Debye approximation** conists in extending the above approximation (that is valid only close to $k = 0$) **to all $\mathbf{k}$ values**. And in considering all the three valuer $c_r$ equal to the same value $c$:
$$\omega_r(\mathbf{k}) = ck\quad \forall r=1,2,3$$
so the three dispersion relations coincide $\forall \mathbf{k}$ in the first Brioullin zone.

##### Density of states $g(\omega)$ in the Debye approximation
Let's define a function $g(\omega)$ such that $g(\omega)d\omega$ gives the number of phonon states with frequency in the range $(\omega, \omega + d\omega)$, per unit volume.

To calculate this function, let's consider how many _avaible states in the Brioullin zone_ have frequency _less than_ $\omega$.
Those states are all the ones where $k < \frac{\omega}{c}$ and, as each avaible state corresponds to the same fraction of volume in reciprocal space ($dV = \frac{(2\pi)^3}{V}$), then the number of avaible states that satisfy the relation is the volume of the reciprocal space that contains them, divided by $dV$.
So we calculate the volume of the sphere with radius $k = \frac{\omega}{c}$, because the number of states with frequency less than $\omega$ is proportional to this volume.
$$N(\omega) = \frac{1}{dV} \frac43\pi\frac{\omega^3}{c^3} = \frac{V}{(2\pi)^3}\frac43\pi\frac{\omega^3}{c^3}$$
where $N(\omega)$ is the number of phonon states with frequency less than $\omega$.
From this result, we can calculate the density of states $g(\omega)$ simply by deriving $N(\omega)$ with respect to $\omega$ and dividing the result by the volume of the crystal $V$. So we find
$$g(\omega)=\frac{1}{2\pi^2}\frac{1}{c^3}\omega^2$$

Some remarks on the above expression
- The result must be multiplyed by 3, as the three bands coincide and the result has been obtained considering only one of those bands.
- The expression is not valid $\forall \omega$. That's because the phonons states are in the Brioullin zone, so the dispersion relation $\omega(\mathbf{k}) = ck$ must be considered **only for** $k_i < \frac{2\pi}{a}$.
So it exists a _cutoff frequency_ $\omega_D$ such that $g(\omega) = 0$ if $\omega > \omega_D$, because those high frequency **do not correspond to any phonon state!**.


You could think that finding the cutoff frequency is as simple as evaluating the dispersion relation $\omega(\mathbf{k}) = ck$ at the border of the Brioullin zone, where the frequency is max.
But the Brioullin zone contains some states that are inside the Brioullin zone, but outside the sphere of radius $k = \frac{\pi}{a}$. This means that the frequencies you find by evaluating the dispersion relation in states **in those corners** are not compatible with the above expression of $g(\omega)$ and neither with $g(\omega)$.

As this complicates the calculations and the Debye approximation is introduced to simplify calculations, we make an additional approximation.

The idea of this last approximation is to modify the shape of the Brioullin zone from a cube into a sphere, but in a way that the number of discrete points $\mathbf{k}$ inside the Brioullin that correspond to avaible states is unchanged, so it remains $N$. In this trasformation, the lattice of spacing $\frac{2\pi}{L}$ of the avaible states $\mathbf{k}$ is left unchanged and so the sphere must contain a number of points of this lattice that is the same of the cube; but what are the contained points can change. 

The advantage you get by changing the shape of the Brioullin zone is that now the isoenergetic surfaces of the dispersion relation $\omega(\mathbf{k}) = ck$ have the same shape (spherical) of the border of the Brioullin zone.
So now, when you look for the cutoff frequency $\omega_D$, **there are no corners!**

As a consequence, the cutoff frequency is determined simply by evaluating the dispersion relation at the border of the brioullin zone. So you can find what's the value of $k$ at the border of the Brioullin zone, that is the one such that the sphere of radius $k$ contains exactly $N$ states $k$. Which is equivalent to ask that the volume of a sphere of radius $k$ is equal to the volume of a cube of side $\frac{2\pi}{a}$, where $a = \frac{L}{N^{\frac13}}$.

Otherwise you can find $\omega_D$ requiring that the number of states of frequency less than $\omega_D$ is the total number of states, which is $3N$, because you have three degenerate bands, so
$$N(\omega_D) = 3N$$
where you can write $$N(\omega_D) = V\int_0^{\omega_D} d\omega g(\omega) = 3N$$

Remember that you need to multiply for V the integral, because $g(\omega)$ is a density of state _per unit volume_.
The two methods for calculating $\omega_D$ are coherent because, in both cases, the result depends on both $N$ and $L$ (we deal with a cubic sample, so $V=L^3$).

Following the first method (that is the less common, but the simpler), if you equate the two volumes and write the radius of the new Brioullin zone as $k = \frac{\omega_D}{c}$,
$$\frac43\pi\frac{\omega_D^3}{c^3} = \frac{(2\pi)^3}{a^3}$$
so, using that $a = \frac{L}{N^{\frac13}}$ you get
$$\omega_D \sim \frac{1}{a} = \frac{N^{\frac13}}{L} = \big(\frac{N}{V}\big)^{\frac13}$$
So
$$\omega_D \sim n^{\frac13}$$
where $n$ is the density of particle per unit volume of the crystal.

### Resumee on the Debye Approximation
$$g(\omega) = \frac{3}{2\pi^2c^3}\omega^2 \sim \omega^2 \quad \text{if} \quad \omega < \omega_D$$
$$\omega_D \sim n^{\frac13}$$
and there are no states for $\omega > \omega_D$, so $g(\omega) = 0$ if $\omega > \omega_D$.

# Total number of phonons in the debye approximation
We want to find the temperature dependence of the total number of phonons in the debye model.

The statistical average of phonons is
$$<N> = \sum_k <n_k>$$
where $$<n_k> = f_{be} = \frac{1}{e^{\beta\hbar\omega(k)} - 1}$$
Note that $\mu = 0$ in a phonon gas.

Introducing the phonons density of states, we have 
$$<N> = \int_0^{\omega_D}d\omega g(\omega) \frac{1}{e^{\beta\hbar\omega} - 1} \sim \int_0^{\omega_D}d\omega \frac{\omega^2}{e^{\beta\hbar\omega} - 1}$$

changing variable into $x = \beta\hbar\omega$ and accounting only the temperature ($\beta$) dependence in the multiplicative coefficient of the integral, we get
$$<N> \sim \frac{1}{\beta^3}\int_0^{\beta\hbar\omega_D}dx\frac{x^2}{e^x - 1}$$

so if we consider the limit of **high temperatures**, so $\beta\rightarrow 0$, we can expand $e^x = 1 + x$ and find
$$<N> \sim \frac{1}{\beta^3}\int_0^{\beta\hbar\omega_D}dx\frac{x^2}{1 + x - 1}$$
so
$$<N> \sim \frac{1}{\beta} = k_bT $$

while at **low temperatures** $\beta\rightarrow\infty$ and so the integral goes from $0$ to $\infty$. This means that the _integral does not depend on temperature_ and so
$$<N>\sim T^3$$

#### Comments on the two results
At low temperatures, the Debye approximation is good, because phonons tend to populate states close to $k = 0$, where the _real_ dispersion law is well approximated by the debye one. And then phonons tend to stay away from the borders of the Brioullin zone.

While at high temperatures, this is no more true.

As a consequence, the result obtained for low temperature in the Debye approximation we expect to be a good extimate of the real value.
While at hight temperatures I don't know if it makes sense trusting this result.
