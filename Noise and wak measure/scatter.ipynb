```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
tmax = 10
sigma = 1
x = np.linspace(0,tmax,N)
y = np.random.normal(0,sigma,N)

fig, ax = plt.subplots()
plt.plot(x,y,'.')
plt.plot(x,np.zeros(N))
plt.plot(x, sigma*np.ones(N),'--',color="red")
plt.plot(x, -sigma*np.ones(N),'--',color="red")
ax.set_xlabel("time t")
ax.set_ylabel("I(t)")
```
