```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
tmax = 10
sigma = 1
x = np.linspace(0,tmax,N)
y = np.random.normal(0,sigma,N)

fig, ax = plt.subplots()
plt.plot(x,y,'.')
plt.plot(x,np.zeros(N))
plt.plot(x, sigma*np.ones(N),'--',color="red")
plt.plot(x, -sigma*np.ones(N),'--',color="red")
ax.set_xlabel("time t")
ax.set_ylabel("Noise source current I(t)")
```

```python
N = int(1e2)
tspan = 5
t = np.linspace(0,tspan,N)
dt = tspan/N
p0 = 1
x0 = 0
ps = p0*np.ones(N)
x = np.zeros(N)
x[0] = x0
p = p0
dp = 1
deltap = np.random.normal(0,dp,N)
for i in np.arange(1,N):
    x[i] = x[i-1] + p*dt
    p = p + deltap[i]  # Heisemberg principle

plt.plot(t,x,label="Backaction")
plt.plot(t,p0*t,label="NO backaction")
plt.xlabel("time t")
plt.ylabel("position x(t)")
plt.title("Misura lenta della posizione\n di una particella libera")
plt.legend()
```

```python
plt.plot(t,ps)
plt.xlabel("time t")
plt.ylabel("momentum $p(t)/p_0$")
plt.title("Misura lenta del momento\n di una particella libera")
```
